```python
def add(n1,n2):
  return n1 + n2
def sub(n1,n2):
  return n1 - n2
def mul(n1,n2):
  return n1 * n2
def div(n1,n2):
  return n1 / n2
```

```python
operations = {
    "+": add,
    "-": sub,
    "*": mul,
    "/": div
}
```

```python
print(operations["*"])
```

```python
print(operations["*"](4,8))
```

```python
def calculator():
  should_accumulate=True
  first_number = float(input("What is the first number?\n"))
  while should_accumulate:
    for symbol in operations:
      print(symbol)
    operation = str(input("pick an operation: "))
    second_number = float(input("What is the second number?\n"))
    ans=operations[operation](first_number,second_number)
    print(f"{first_number} {operation} {second_number} = {ans}")
    choice = str(input("Would you like to continue calculating with your number?\nYes (y) or No (n) ? \n"))

    if choice == "y":
      first_number = ans
    else:
      should_accumulate = False
      #calculator()
  else:
      print("Goodbye")

calculator()
```
