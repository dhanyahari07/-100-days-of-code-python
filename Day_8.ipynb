```python
import string

# Get all lowercase letters as a string
lowercase_str = string.ascii_lowercase  # 'abcdefghijklmnopqrstuvwxyz'

# Convert to a list of characters
lowercase_list = list(lowercase_str)

print(lowercase_list)
```

```python

direction = input("type 'encode ' to encrypt, type 'decode' to decrypt:\n").lower()
plain_text = input (" Type your message:\n").lower()
shift = int(input("Type the shift number:\n"))
```

```python
def encrypt(plain_text,shift_amount):
  cipher_text = ""
  for letter in plain_text:
    shifted_position = lowercase_list.index(letter)+shift_amount
    shifted_position%= len(lowercase_list)
    cipher_text += lowercase_list[shifted_position]
  print(f"The encoded text is {cipher_text}")
```

```python
def decrypt(plain_text,shift_amount):
  cipher_text = ""
  for letter in plain_text:
    shifted_position = lowercase_list.index(letter)-shift_amount
    shifted_position%= len(lowercase_list)
    cipher_text += lowercase_list[shifted_position]
  print(f"The encoded text is {cipher_text}")
```

```python
encrypt(plain_text=plain_text,shift_amount=shift)
#
```

```python
decrypt(plain_text=plain_text,shift_amount=shift)
```

```python
def ceaser(plain_text,shift_amount,direction):
  cipher_text = ""
  if direction == "decode":
    shift_amount *= -1
  for letter in plain_text:
    if letter not in lowercase_list:
      cipher_text += letter
    else:
      shifted_position = lowercase_list.index(letter)+shift_amount
      shifted_position %= len(lowercase_list)
      cipher_text += lowercase_list[shifted_position]
  print(f"The {direction}d text is {cipher_text}")
```

```python
ceaser(plain_text=plain_text,shift_amount=shift,direction=direction)
```

```python
should_continue=True

while should_continue:
  direction = input("type 'encode ' to encrypt, type 'decode' to decrypt:\n").lower()
  plain_text = input (" Type your message:\n").lower()
  shift = int(input("Type the shift number:\n"))
  ceaser(plain_text=plain_text,shift_amount=shift,direction=direction)
  restart=input("type 'yes' if you want to go again. Otherwise type 'no'.\n").lower()
  if restart == "no":
    should_continue = False
    print("Goodbye")

```
