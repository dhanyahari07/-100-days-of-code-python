```python
import random
```

```python
# List of hangman stages from empty to fully drawn
HANGMAN_STAGES = [
    """
       +---+
       |   |
           |
           |
           |
           |
    =========
    """,
    """
       +---+
       |   |
       O   |
           |
           |
           |
    =========
    """,
    """
       +---+
       |   |
       O   |
       |   |
           |
           |
    =========
    """,
    """
       +---+
       |   |
       O   |
      /|   |
           |
           |
    =========
    """,
    """
       +---+
       |   |
       O   |
      /|\\  |
           |
           |
    =========
    """,
    """
       +---+
       |   |
       O   |
      /|\\  |
      /    |
           |
    =========
    """,
    """
       +---+
       |   |
       O   |
      /|\\  |
      / \\  |
           |
    =========
    """
]

```

```python
word_list=["aardvark","baboon","camel"]

chosen_word=random.choice(word_list)

print(chosen_word)
```

```python
placeholder=""
word_length=len(chosen_word)
for i in range(word_length):
    placeholder+="_"
print(placeholder)
```

```python
#guess=input("guess a letter").lower()

game_over=False
correct_letters = []
lives=6

while not game_over:
  guess=input("guess a letter").lower()

  display=""

  for letter in chosen_word:
    if letter==guess:
      display+=letter
      correct_letters.append(letter)
    elif letter in correct_letters:
      display+=letter
      #print("right")
    else:
      display+="_"
      #print("wrong")
  print(display)

  if guess not in chosen_word:
    lives-=1
    if lives==0:
      game_over=True
      print("you lose")
  if "_" not in display:
    game_over=True
    print("you win")

  print(HANGMAN_STAGES[6-lives])
```
